# Task
Create three dataframes from the provided data for `patients.csv`, `treatments.csv`, and `outcomes.csv`, merge them, and then perform the following analyses: group by `treatment_type` to find the average recovery days and success rate; pivot the data to show each patient's total cost per treatment type; find which treatment type has the highest average cost; find which patient had the longest recovery time; and analyze the success rate by gender.

## Create dataframes

### Subtask:
Create pandas DataFrames from the provided data for `patients.csv`, `treatments.csv`, and `outcomes.csv`.


**Reasoning**:
Import pandas and load the three CSV files into separate dataframes as instructed.



```python
import pandas as pd

# Create dataframes from the provided data
patients_df = pd.DataFrame({
    'patient_id': ['P01', 'P02', 'P03'],
    'name': ['John', 'Maya', 'Alice'],
    'age': [45, 30, 55],
    'gender': ['M', 'F', 'F']
})

treatments_df = pd.DataFrame({
    'treatment_id': ['T101', 'T102', 'T103'],
    'patient_id': ['P01', 'P02', 'P01'],
    'treatment_type': ['Chemotherapy', 'Physiotherapy', 'Surgery'],
    'cost': [5000, 1500, 8000]
})

outcomes_df = pd.DataFrame({
    'treatment_id': ['T101', 'T102', 'T103'],
    'recovery_days': [30, 10, 60],
    'success': ['Yes', 'Yes', 'No']
})

print("Patients DataFrame:")
display(patients_df)
print("\nTreatments DataFrame:")
display(treatments_df)
print("\nOutcomes DataFrame:")
display(outcomes_df)
```

```python
# Merge treatments and outcomes dataframes on 'treatment_id'
merged_df = pd.merge(treatments_df, outcomes_df, on='treatment_id')

# Merge the combined dataframe with patients dataframe on 'patient_id'
merged_df = pd.merge(merged_df, patients_df, on='patient_id')

print("\nMerged DataFrame:")
display(merged_df)
```

```python
# Convert 'success' column to numerical (1 for Yes, 0 for No)
merged_df['success_numerical'] = merged_df['success'].apply(lambda x: 1 if x == 'Yes' else 0)

# Group by treatment_type and find the average recovery days and success rate
treatment_analysis = merged_df.groupby('treatment_type').agg(
    average_recovery_days=('recovery_days', 'mean'),
    success_rate=('success_numerical', 'mean')
)

print("\nAnalysis by Treatment Type:")
display(treatment_analysis)
```

```python
# Find which treatment type has the highest average cost
average_cost_by_treatment = merged_df.groupby('treatment_type')['cost'].mean().sort_values(ascending=False)

print("\nAverage Cost by Treatment Type:")
display(average_cost_by_treatment)

highest_average_cost_treatment = average_cost_by_treatment.index[0]
print(f"\nTreatment type with the highest average cost: {highest_average_cost_treatment}")
```
